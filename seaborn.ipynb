## Data visualaization using matplot and seaborn 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
print("libraries imported!")
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# sns.lineplot()
# sns.barplot()
# sns.boxplot()
# sns.scatterplot()
# sns.histplot()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
df.head(1)
```

```python
sns.relplot(y='tip',x='total_bill',data=df,hue='sex')
plt.show()
```

```python
df.head(1)
```

```python
sns.relplot(y='tip',x='total_bill',data=df,hue='sex',style='time')
plt.show()
```

```python
sns.relplot(y='tip',x='total_bill',data=df,hue='sex',style='time',col='smoker')
plt.show()
```

```python
sns.relplot(y='tip',x='total_bill',data=df,hue='time',style='sex',col='day')
plt.show()
```

```python
# scatterplot , lineplot , kind = 'scatter'
sns.relplot(y='tip',x='total_bill',data=df,hue='time',style='sex',col='day',kind='line')
plt.show()
```

```python
df.head(1)
```

#### categorical plot catplot()<br>

    "boxen", "point", "bar", or "count".

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='bar')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='violin')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='box')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='boxen')
plt.show()
```
